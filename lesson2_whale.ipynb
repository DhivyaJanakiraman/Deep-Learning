```python
from bing_image_downloader import downloader
```

```python
for q in ["killer whale", "beluga whale", "humpback whale"]:
    downloader.download(q, limit=20, output_dir='whales', adult_filter_off=True, force_replace=False, timeout=5)
```

```python
path = Path('whales')
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```

```python
whales = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = whales.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
whales = whales.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls = whales.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
whales = whales.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'))
dls = whales.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
whales = whales.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = whales.dataloaders(path,bs=5)
dls.train.show_batch(max_n=4, nrows=1, unique=True)
```

```python
whales = whales.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = whales.dataloaders(path, bs=5)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```python
whales = whales.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = whales.dataloaders(path, bs=5)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```python
whales = whales.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = whales.dataloaders(path, bs=5)
```

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
learn.export()
```

```python
btn_upload = widgets.FileUpload()
btn_upload
```

```python
btn_upload = SimpleNamespace(data = ['whales/killer whale/Image_10.jpg'])
```

```python
img = PILImage.create(btn_upload.data[-1])
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
out_pl
```

```python
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
lbl_pred
```

```python
btn_run = widgets.Button(description='Classify')
btn_run
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
btn_upload = widgets.FileUpload()
```

```python
VBox([widgets.Label('Select your whale!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
